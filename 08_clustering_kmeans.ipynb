```python
def Draw(pred, features, poi, mark_poi=False, name="image.png", f1_name="feature 1", f2_name="feature 2"):
    ### traza cada grupo con un color diferente; agrega más colores para
    ### dibujar más de cinco grupos
    colors = ["b", "c", "k", "m", "g"]
    for ii, pp in enumerate(pred):
        plt.scatter(features[ii][0], features[ii][1], color = colors[pred[ii]])

    ### si mark_poi es True, coloca estrellas rojas sobre los puntos que son puntos de interés
    if mark_poi:
        for ii, pp in enumerate(pred):
            if poi[ii]:
                plt.scatter(features[ii][0], features[ii][1], color="r", marker="*")

    plt.xlabel(f1_name)
    plt.ylabel(f2_name)
    #plt.savefig(name)
    plt.show()
```

```python
### cargar el diccionario que contiene los datos de cada persona en el conjunto de datos
data_dict = joblib.load( open("./tools/final_project_dataset.pkl", "rb") )

### Remover outlier
data_dict.pop("TOTAL", 0)

### funciones de entrada que queremos usar
### puede ser cualquier clave en el diccionario a nivel de persona
### (salario, honorarios de director, etc.)
feature_1 = "salary"
feature_2 = "exercised_stock_options"
feature_3 = "total_payments"
poi  = "poi"
features_list = [poi, feature_1, feature_2, feature_3]
data = featureFormat(data_dict, features_list )
poi, finance_features = targetFeatureSplit( data )

### Visualizar
for f1, f2, f3 in finance_features:
    plt.scatter(f1, f2)

plt.xlabel(feature_1)
plt.ylabel(feature_2)
plt.show()
```

```python
### Aquí se ejecuta el algoritmo de clustering Kmeans
### Crear predicciones de las etiquetas del grupo
### para los datos y almacenarlos en una lista llamada pred

from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler

est = KMeans(n_clusters=2, n_init=10)
est.fit(finance_features)
pred = est.predict(finance_features)

try:
    Draw(pred, finance_features, poi, mark_poi=False, f1_name=feature_1, f2_name=feature_2)
except NameError:
    print("No se encontró ningún objeto de predicciones llamado pred, no hay grupos para graficar.")
```
